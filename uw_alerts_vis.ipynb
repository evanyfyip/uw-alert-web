```python
import pandas as pd
import re
```

```python
file_name = "./data/UW_Alerts_2018_2022.txt"
with open(file_name) as f:
    lines = f.readlines()
    alerts = []
    alert = ''
    for line in lines:
        if line != '\n':
            line_list = line.split()
            if bool(re.search('^[a-zA-Z]+\s\d{1,2},\s\d{4}$', line)): # Searching for the date
                alerts.append(alert)
                alert = ''
                alert += line
            else:
                alert += line
```

```python
x = 0
uw_alerts = []
for alert in alerts:
    if bool(re.search('UW Alert Seattle.*', alert)) | bool(re.search('shooting', alert)) | bool(re.search('robbery', alert)):
        x += 1
#         print(alert)
#         print("+--------------------------------------------------------------+")
        uw_alerts.append(alert)
```

```python
uw_alerts
```

```python
original_posts = []
dates = []
updates = []
for alert in uw_alerts:
    # Searching for updates
    alert_updates = re.findall('UPDATE.*\n', alert)
    updates.append(alert_updates)
    
    # Searching for dates
    match = re.search('^[a-zA-Z]+\s\d{1,2},\s\d{4}', alert)
    if match is not None:
        span = match.span()
        dates.append(alert[span[0]:span[1]])
    # Extracting original post
    match = re.search('.*ORIGINAL POST.*\n.*', alert)
    if match is not None:
        span = match.span()
#         print(span)
#         print(match.span())
        original_posts.append(alert[span[0]:span[1]])
    else:
        original_posts.append(None)
```

```python
alerts_df = pd.DataFrame({"date":dates, "alert":uw_alerts, "original_post":original_posts, "updates":updates})
alerts_df['time'] = alerts_df["original_post"].str.extract(r'(\d{1,2}:\d{1,2}\s[a,p,A,P].[m, M].)')
alerts_df['time_2'] = alerts_df["original_post"].str.extract(r'(\d{1,2}:\d{1,2}[a,p,A,P][m,M])')
alerts_df['time_3'] = alerts_df["original_post"].str.extract(r'(\s\d{1,2}[a,p][m])')
alerts_df['time_4'] = alerts_df["original_post"].str.extract(r'(\d{1,2}:\d{1,2}\s[a,A,p,P][m,M])')
```
